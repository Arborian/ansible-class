# Exploring EC2 and Ansible

<http://docs.ansible.com/ansible/latest/list_of_cloud_modules.html#amazon>

Many of our commands with EC2 will use host delegation (particularly running locally)

By default, it will use the AWS CLI configuration

````python
%%bash
ansible localhost -m ec2_instance_facts
````

We can also, of course, launch instances with the `ec2` module, configure ELBs with the `ec2_elb` module, and much, much more. One thing of interest that we can do is launch spot instances with a bid:

```
- ec2:
    spot_price: 0.24
    spot_wait_timeout: 600
    keypair: mykey
    group_id: sg-1dc53f72
    instance_type: m1.small
    image: ami-6e649707
    wait: yes
    vpc_subnet_id: subnet-29e63245
    assign_public_ip: yes
    spot_launch_group: report_generators
```


We can touch almost all the AWS services, including of course S3:

````python
%%bash
ansible localhost -m aws_s3_bucket_facts
````
